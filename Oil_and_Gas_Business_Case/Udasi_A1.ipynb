## Problem Statement

The core business challenge lies in effectively predicting the future productivity of oil and gas wells, the focus is on identifying the characteristics that make a "good" well for the purpose of making informed decisions on purchasing production rights. This challenge encompasses:

- Current forecasting methods rely on manually created ‘type curves’ that approximate future production based on historical data of similar wells.  

- The inherently uncertain nature of oil and gas production, influenced by fluctuating market prices, regulatory changes, and technical challenges, complicates the investment decision-making process.

The goal is to maximize return on investment by identifying wells most likely to yield profitable returns once the oil or gas is extracted and sold. Enhancing the accuracy of these forecasts is crucial for reducing investment risks and maximizing returns. Misjudgments in this area can lead to significant financial losses and missed opportunities for profitable investments.

The company has access to extensive datasets detailing the metadata and historical production data of oil and gas wells in New Mexico, Permian region of the Delaware basin, focusing on wells with horizontal trajectories. Integrating and analyzing this data to identify the characteristics of potentially profitable wells is a complex task that requires  analytical approaches. The problem should be approached by identifying and creating relevant features from the available data that can predict well productivity and by analyzing historical production volumes, production rates, and decline rates.

### Sub-Problem statement 1 - Impact of the lateral length of the well on its performance

The first sub-problem focuses on identifying the impact of the lateral length of the well on its performance. Specifically, it aims to analyze how variations in lateral length influence key performance metrics such as EURWH_MBOE (Estimated Ultimate Recovery for Water and Hydrocarbons in Thousand Barrels of Oil Equivalent), GOR_ScfPerBbl (Gas Oil Ratio in Standard Cubic Feet Per Barrel), TotalOpex_USDPerBOE (Total Operational Expenditure in USD Per Barrel of Oil Equivalent), and CapitalEfficiency_USDPerBOEPerDAY (Capital Efficiency in USD Per Barrel of Oil Equivalent Per Day). 

Understanding these relationships is crucial for assessing the financial viability and operational efficiency of acquiring production rights for different wells. This analysis is vital for several reasons:

- Investment Decision-Making: It directly informs the investment decision-making process by identifying the most profitable well characteristics, enabling the company to prioritize investments in wells with the most favorable performance metrics.
- Resource Allocation: Insights from this analysis help in optimal resource allocation, ensuring that capital and operational efforts are directed towards wells with the highest return on investment.


### Sub-Problem statement 2 - Understand how water production changes over the useful life of a well, as oil and gas production depletes.

The second sub-problem seeks to understand how water production changes over the useful life of a well, as oil and gas production depletes. Given that disposing of contaminated water can become a significant expense, this analysis focuses on creating a Cumulative Water to Oil Ratio (Cum_Water_to_Oil) and grouping wells by Production Stage (e.g., '0-1 year', '1-5 years', '5-10 years', '10+ years') to assess the financial and operational impacts of water production on well profitability over time.

Understanding that high water production can lead to increased disposal costs, impacting the overall profitability of a well. This consideration is crucial for effective cost management and comparing targets.

### Sub-Problem statement 3 - Analyzing historical production rates of oil and gas wells based on Lateral Lengths and Production Stages.
The third sub-problem focuses on analyzing historical production rates of oil and gas wells based on Lateral Lengths and Production Stages. This involves the application of type curve analysis to estimate the rate at which well production decreases over time. The objective is to refine the process of generating type curves, which graphically represent production rate versus time, based on historical production data from similar wells. 

Understanding of these type curves are essential for evaluating the value of production rights being traded, thus aiding in strategic investment, divestment, and portfolio management decisions about purchasing production rights.

# Assumptions made

## Problem 1

- **Uniformity Across Wells**: It is assumed that the categorization of wells based on lateral length is representative of similar operational conditions and geological characteristics. This assumes that all wells within a category are comparable and that lateral length is a significant independent variable influencing the KPIs.
- **Median as a Representative Statistic**: The use of median values to represent the central tendency of the data assumes that the median is an adequate measure of central tendency for the distribution of each KPI, even though this may not account for skewed distributions or outliers.
- **Stable Production Costs**: By calculating the median of Total Operational Expenditure (TotalOpex_USDPerBOE) and Capital Efficiency (CapitalEfficiency_USDPerBOEPerDAY), there is an underlying assumption that cost factors are relatively stable over time and across different well operations.

## Problem 2
- **Median as an Indicator**: It is assumed that the median values of the metrics are representative of typical well performance within each category. This assumes that the data is not significantly skewed, and that outliers do not disproportionately influence the well performance metrics.
- **Consistent Reporting**: The analysis assumes that all operators report production and operational metrics consistently and accurately. This means that the Capital Efficiency, and other KPIs are based on similar definitions and calculation methods across the dataset.
- **Stable Technological and Operational Conditions**: The pivot table analysis presumes that the technological and operational conditions under which the wells operate remain constant over time. It assumes that advancements in technology or changes in operational techniques have not significantly impacted the performance metrics across different production stages or lateral lengths.

## Problem 3

The water-to-BOE ratio can be influenced by many factors, including the maturity of wells, the geological formation, and the method of production.

- **Consistency in Decline Rates**: It is assumed that historical decline rates, as observed in the provided data, will continue into the future, allowing the type curves to be predictive. This assumes that the geological and operational conditions remain consistent over time for similar wells.
- **Representativeness of the Sample**: The data used to generate the type curves is assumed to be representative of the wider population of oil and gas wells. This means the wells in the dataset are considered a good proxy for similar wells not included in the analysis.
- **Technology and Practices Stagnation**: The analysis presupposes that there have been no significant technological or operational advancements that could alter production efficiency. This means that the impact of such advancements on production rates is not factored into the type curves.
- **Consistency of Historical Trends**: It is assumed that past production trends are reliable indicators of future performance. This implies that historical cumulative production data used to create type curves will continue to accurately forecast future production rates.
- **Cost-Profit Balance**: The analysis assumes that the additional costs incurred from drilling longer laterals are balanced by the increased extraction and sale of oil and gas resources. This is based on the idea that the higher initial investment will be justified by the greater cumulative production over time.
- **Stable Market Conditions**: It is presumed that market conditions, including the selling price of oil and gas, will remain favorable enough to ensure that the costs of drilling and operating longer lateral wells are recoverable and profitable over the well's productive life.

## Problem 1

- **Simplicity of the Model**: Simplifying the analysis to a few KPIs may overlook the complexity of oil and gas production. For example, it does not consider the multi-dimensional interplay between lateral length, well depth, drilling techniques, and reservoir properties, which all can significantly impact well performance.

## Problem 2

- **Sample Size**: The number of wells in the '10+ years' category is very low (3 wells), which may not provide a reliable representation of this stage.
- **Extrapolation**: The results may not be universally applicable across all geological formations or operational conditions.
- **Temporal Factors**: The analysis does not account for changes over time that could affect well performance, such as depletion of reserves, regulatory changes, market conditions, or advancements in recovery technologies.
- **Lateral Length as a Sole Factor**: Focusing primarily on lateral length may overlook other influential factors such as well depth, geological variations, and completion techniques that can also significantly impact well performance. This could lead to overestimation or underestimation of the influence of lateral length on the performance metrics.

## Problem 3

- **Market and Regulatory Changes**: The analysis does not account for potential market fluctuations or regulatory changes that could significantly affect future production rates and the economic viability of the wells.
- **Data Scope**: If the dataset predominantly contains wells from a specific region or managed by certain operators, the findings may not be generalizable to wells in different regions or with different management practices.
- **Unaccounted Variables**: The analysis primarily focuses on lateral length and production stage without considering other potentially influential factors such as reservoir characteristics, completion techniques, or well maintenance practices that could affect production rates and the applicability of type curves.
- **Changing Technological Landscape**: The analysis may not fully account for future advancements in drilling and extraction technology that could alter the cost structures or production efficiencies of wells, particularly those with longer laterals.
- **Variability in Geology**: There is an inherent variability in subsurface geology that can impact production rates and costs, which may not be fully captured by the analysis. This means that the recommendations might not hold true for all geographic locations or formations.
- **Operational Risks**: The recommendation does not consider potential operational risks and uncertainties that can affect production, such as equipment failures, unexpected maintenance requirements, and environmental hazards, which could disproportionately affect longer lateral wells due to their complexity.

```python
# Function to plot a correlation heatmap with specified parameters
# Parameters:
#   - corr: correlation matrix
#   - threshold: correlation threshold to display
#   - mask_upper_triangle: boolean flag to mask the upper triangle of the heatmap
def plot_correlation_heatmap(corr, threshold=0.5, mask_upper_triangle=False):
    """
    Function to plot a correlation heatmap with specified parameters
    
    Parameters:
    corr : pandas DataFrame
        Correlation matrix
    threshold : float, optional
        Correlation threshold to display (default is 0.5)
    mask_upper_triangle : bool, optional
        Flag to mask the upper triangle of the heatmap (default is False)
    """
    plt.figure(figsize=(12, 10))
    
    # Generate a mask for the upper triangle
    if mask_upper_triangle:
        mask = np.triu(np.ones_like(corr, dtype=bool))
    else:
        mask = None

    # Only show the strong correlations
    sns.heatmap(corr[(corr >= threshold) | (corr <= -threshold)],
                cmap='viridis',
                mask=mask,
                vmax=1.0,
                vmin=-1.0,
                linewidths=0.1,
                annot=True,
                annot_kws={"size": 10},
                square=True)
```

```python
# Calculate the mean cumulative oil production for each 'RawOperator' and 'TotalProdMonths'
LateralCategory_cumprod = production_df.groupby(['LateralCategory', 'TotalProdMonths'])['CumProd_BOE'].mean()

# Since you want cumulative sums over 'TotalProdMonths' for each operator, ensure the data is sorted
LateralCategory_cumprod = LateralCategory_cumprod.sort_index(level=[0, 1])

plt.figure(figsize=(12, 8))

# Plotting each lateral category's trend
for LateralCategory, data in LateralCategory_cumprod.groupby(level=0):
    plt.plot(data.index.get_level_values(1), data.values, label=LateralCategory, linestyle='-')

plt.xlabel('Total Production Months')
plt.ylabel('Mean Cumulative Production (BOE)')
plt.title('Cumulative Oil & Gas Production Over Time by Lateral Category')
plt.grid(True)
plt.xticks(rotation=45)

# Adjusting the legend to be outside the plot area to accommodate multiple entries
plt.legend(loc='upper left', fontsize='small', bbox_to_anchor=(1, 1))
plt.tight_layout()

# Show the plot
plt.show()
```
