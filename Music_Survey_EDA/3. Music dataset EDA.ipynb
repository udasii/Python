# Traversing the DataFrame
In the vast sea of data within a dataframe, finding the information you need can be like searching for a needle in a haystack, especially with lengthy tables. Fortunately, pandas provides an array of methods to streamline this process. Let's explore some of the most prevalent and valuable techniques for extracting the desired information.

## `loc`

- Purpose: Accesses a group of rows and columns by labels or boolean array.
- Usage: `df.loc[row_label, column_label]`
- Useful When: You know the label of the index or columns.

```python
# Access the data for the third respondent
third_respondent = survey_df.loc[2]
third_respondent
```

The output is the third register in the df (remember in Python we start at zero). Since the index was just a number we could access it this way, but in other dfs we will have to put the index value directly, like we do in the following cell with the column name `Fav genre`

```python
third_respondent_fav_genre = survey_df.loc[2,'Fav genre']
third_respondent_fav_genre
```

## `iloc`
- Purpose: Accesses a group of rows and columns by integer position(s).
- Usage: `df.iloc[row_index, column_index]`
- Useful When: You want to access elements based on their position (integer index).

```python
# Access the data for the a slice of the second to the fifth register and second to sixth coulmn
second_to_fifth = survey_df.iloc[2:5,2:6]
second_to_fifth
```

The operation we performed in the previous cell is called a **slicing**. We specify ranges of rows and columns we want to have. This is useful when applying operations only to those specific rows and columns

## `isin`
- Purpose: Filter data based on a list of values.
- Usage: `df[column].isin(values)`
- Useful When: Filtering data by a list of categorical values.

```python
# Select respondents who use specific streaming services
services = ['Spotify', 'Apple Music']
filtered_df = survey_df[survey_df['Primary streaming service'].
                        isin(services)]
filtered_df
```

## Masking
- Purpose: Selecting data based on a condition.
- Usage: `df[mask]` where mask is a boolean array.
- Useful When: You need to filter data based on a specific condition.

```python
# Get a boolean pd.Series respondents aged over 30
over_30 = survey_df['Age'] > 30
over_30
```

Now we can select only the values where the boolean series is True by using the following notation

`df_masked=df[mask]`

note that both the mask and the df will have to have the same indexes.

```python
# Select respondents aged over 30
df_over_30 = survey_df[survey_df['Age'] > 30]
df_over_30
```

When masking multiple conditions you need to wrap each of them in a parenthesis and the logic operators change form `and` to `&` and from `or` to `|`. As we can see in the following example

```python
streaming_services=['Pandora', 'YouTube Music']
df_over_30_pandora_yt = survey_df[(survey_df['Age'] > 30) & 
    (survey_df['Primary streaming service'].isin(streaming_services))]
df_over_30_pandora_yt.head()
```

## Querying:
- Purpose: Query the columns of a DataFrame with a boolean expression.
- Usage: `df.query('expression')`
- Useful When: You want to filter data using a concise, readable syntax.

```python
# Select respondents who spend more than 3 hours per day and are over 25 years old
specific_group = survey_df.query('`Hours per day` > 3 & Age > 25')
specific_group
```

Note that we could have achieved the same thing using masking. The differences between these two methods are shown in the table below

| Aspect                      | Querying                            | Masking                             |
|-----------------------------|-------------------------------------|-------------------------------------|
| **Usage Scenario**          | When you need to filter data using a concise, readable string expression. Ideal for complex conditions. | When you want to filter data based on conditions directly written with DataFrame's columns. More Pythonic for simple conditions. |
| **Syntax Example**          | `df.query('Age > 30 & "Hours per day" > 3')` | `df[(df.Age > 30) & (df['Hours per day'] > 3)]` |
| **Computational Efficiency**| Generally faster for complex queries due to optimized internal parsing and execution. | Slightly slower for complex conditions due to direct evaluation in Python's interpreter. However, differences are often minimal for small to moderately sized DataFrames. |
| **In-Place / Out-of-Place** | Out-of-place. Returns a new DataFrame. | Out-of-place. Returns a new DataFrame. |
| **Readability**             | Better for complex conditions. Compact and closer to SQL-like syntax. | Better for simple conditions. Follows standard Python logical syntax. |
| **Flexibility**             | Limited to string expressions. Cannot use external variables directly without formatting. | More flexible. Allows the use of external Python variables and functions directly. |
| **Additional Notes**        | Requires string parsing. Might throw an error for invalid syntax. | Utilizes standard Python boolean indexing. Familiar to Python users. |
